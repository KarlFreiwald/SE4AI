# Computing initialization images

```python
import numpy as np
import matplotlib.pyplot as plt

from helpers import load_data_from_pickle, save_data_pickle
```

```python
# Replace with your own database download directory
# base_path = 'E:/input'
base_path = r'C:\Users\kfrei\OneDrive - Ostbayerische Technische Hochschule Regensburg\Desktop\input'
```

```python
pickle_index = 2
x_train, y_train, x_val, y_val, x_test, y_test = load_data_from_pickle(index=pickle_index)
```

## Computing the Average Images

```python
# Compute the average images
avg_image_train = np.mean(x_train, axis=0)
avg_image_val = np.mean(x_val, axis=0)
avg_image_test = np.mean(x_test, axis=0)
```

```python
# Saving the average images as pickle files
save_data_pickle(avg_image_train, 'average_images/avg_image_train.pickle')
save_data_pickle(avg_image_val, 'average_images/avg_image_val.pickle')
save_data_pickle(avg_image_test, 'average_images/avg_image_test.pickle')
```

```python
avg_image_train.shape
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow((avg_image_train * 255).astype('uint8'))
axs[0].set_title('Average Image of x_train')
axs[0].axis('off')

axs[1].imshow((avg_image_val * 255).astype('uint8'))
axs[1].set_title('Average Image of x_val')
axs[1].axis('off')

axs[2].imshow((avg_image_test * 255).astype('uint8'))
axs[2].set_title('Average Image of x_test')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
